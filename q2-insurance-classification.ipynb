```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("../input/insurance-dataset/Insurance cross sell.csv")
data.head()
```

```python
data.shape
```

```python
data.dropna()
```

```python
data.shape
```

```python
data.describe()
```

```python
data.drop('id',axis=1,inplace=True)
```

```python
data.head()
```

```python
gender=pd.get_dummies(data['Gender'],drop_first=True)
```

```python
vehicle_age=pd.get_dummies(data['Vehicle_Age'],drop_first=True)
```

```python
vehicle_damage=pd.get_dummies(data['Vehicle_Damage'],drop_first=True)
```

```python
data= pd.concat([data,gender,vehicle_age,vehicle_damage],axis=1)
```

```python
data.head()
```

```python
data.drop(['Gender','Vehicle_Age','Vehicle_Damage'],axis=1,inplace=True)
```

```python
data.head()
```

```python
X= data.drop('Response',axis=1)
Y= data['Response']
```

```python
from sklearn.model_selection  import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=101)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(max_iter=1000).fit(X_train, Y_train)
```

```python
pred1 = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
test_score = accuracy_score(Y_test,pred1)
```

```python
print(test_score)
```

K-NEAREST NEIGHBOUR

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=5)
```

```python
neigh.fit(X_train, Y_train)
```

```python
pred2 = neigh.predict(X_test)
```

```python
test_score = accuracy_score(Y_test,pred2)
```

```python
print(test_score)
```

```python
gnb.fit(X_train, Y_train)
```

```python
pred3=gnb.predict(X_test)
```

```python
test_score = accuracy_score(Y_test,pred3)
```

```python
print(test_score)
```
